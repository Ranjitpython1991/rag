1️⃣ What are Embeddings?

Embeddings are numerical vector representations of text that capture the semantic meaning of the text.

They convert words, sentences, or documents into numbers so that machines can understand similarity.


## Similarity metrics

Several metrics are commonly used to compare embeddings:
    
Cosine similarity — measures the angle between two vectors.

Euclidean distance — measures the straight-line distance between points.

Dot product — measures how much one vector projects onto another.

Here’s an example of computing cosine similarity between two vectors:

https://docs.langchain.com/oss/python/integrations/embeddings#top-integrations

```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
```

```python

loader = TextLoader(r"doc/data.txt")
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)

chunks = splitter.split_documents(docs)
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
#embed single texts or documents with embed_query
text = "LangChain is the framework for building context-aware reasoning applications"
single_vector = embeddings.embed_query(text)
print(str(single_vector)[:100])  # Show the first 100 characters of the vector
```

```python
#multiple texts
text2 = (
    "LangGraph is a library for building stateful, multi-actor applications with LLMs"
)
two_vectors = embeddings.embed_documents([text, text2])
for vector in two_vectors:
    print(str(vector)[:100])  # Show the first 100 characters of the vector
```

#  Vector store 

A vector stores embedded data and performs similarity search.

1 add_documents - Add documents to the store.

2 delete - Remove stored documents by ID.

3 similarity_search - Query for semantically similar documents.


Vector stores search using semantic similarity

Types of Vector Stores

1️⃣ In-Memory Vector Stores -  Data is stored in RAM.

Pros:

✔ Very fast
✔ Simple to use

Cons:

❌ Data lost when program stops

Example:

FAISS

2️⃣ Persistent Vector Stores

Data is stored on disk or server.

Pros:

✔ Data persists
✔ Suitable for production

Cons:

❌ Slightly slower than in-memory

Examples:

Chroma

Pinecone

Weaviate

Milvus

```python
from langchain_core.vectorstores import InMemoryVectorStore
vector_store = InMemoryVectorStore(embedding=embeddings)
vector_store.add_documents(chunks)
```

# Similarity search
Issue a semantic query using similarity_search, which returns 
the closest embedded documents:
Many vector stores support parameters like:
k — number of results to return
filter — conditional filtering based on metadata

Metadata filtering
Filtering by metadata (e.g., source, date) can refine search results:
vector_store.similarity_search(
  "query",
  k=3,
  filter={"source": "tweets"}
)



```python
results = vector_store.similarity_search("plant useful ", k=1)
print(results[0].page_content)
```

1️⃣ Similarity

Similarity refers to the mathematical comparison between vectors.

When text is converted into embeddings (vectors), the system compares vectors to find which ones are closest.

Common similarity methods:

Cosine similarity

Euclidean distance

Dot product

📌 Purpose:
Find which stored vectors are most similar to the query vector.

Example flow:

User Query → Convert to Embedding → Compare with stored embeddings → Return top similar vectors

2️⃣ Retriever (get_relevant_documents)

A Retriever is a higher-level component that uses similarity search internally.

Functions like:

get_relevant_documents()

invoke()

similarity_search()

are used to retrieve documents for the LLM.

📌 Purpose:
Return the actual documents or chunks that are relevant to the query.

```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")


loader = TextLoader(r"doc/data.txt")
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)

chunks = splitter.split_documents(docs)
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
# importing vector stores
from langchain_core.vectorstores import InMemoryVectorStore
vector_store = InMemoryVectorStore(embedding=embeddings)
vector_store.add_documents(chunks)
results = vector_store.similarity_search("plant useful ", k=1)
print(results[0].page_content)
```
