Retrieval

Retrieval means fetching relevant information from a knowledge source (documents, database, vector store) before sending it to an LLM to generate an answer.

2. Why Retrieval is Important

Large Language Models like GPT-4, Claude, or LLaMA have limitations:

# Problems Without Retrieval

Knowledge cutoff

Hallucinations

Cannot access private data

Cannot update knowledge easily

# Retrieval Solves These Problems

Provides latest data

Reduces hallucination

Enables enterprise knowledge search

Makes LLM responses grounded in real documents

| Feature    | Similarity                                                    | Retriever                                               |
| ---------- | ------------------------------------------------------------- | ------------------------------------------------------- |
| Definition | A method used to measure how similar two vectors or texts are | A component that returns relevant documents for a query |
| Role       | Mathematical calculation                                      | Search interface                                        |
| Function   | Calculates closeness between vectors                          | Uses similarity to find documents                       |
| Output     | Similarity score                                              | Relevant document chunks                                |
| Used in    | Vector search                                                 | RAG pipelines                                           |


similarity_search() → directly searches similar chunks in the vector store

as_retriever() → converts the vector store into a retriever interface

invoke() → sends a query to the retriever to get relevant documents

# 1.  similarity_search()

Purpose:
Directly search the vector database and return similar documents.

It belongs to the vector store.

Example vector stores:

FAISS

Chroma

Pinecone

Example
results = vector_store.similarity_search("plant useful", k=1)


for doc in results:
    print(doc.page_content)
What happens internally

Query
  ↓
Embedding
  ↓
Similarity calculation
  ↓
Return top-k chunks
Key point

similarity_search() is direct vector database search.

# 2.  as_retriever()

Purpose:
Convert a vector store into a retriever object.

This is used in Retrieval-Augmented Generation pipelines.

Example:

retriever = vector_store.as_retriever()

Now the vector store behaves like a retriever interface.

Why?

Because RAG pipelines expect a retriever object, not a raw vector store.

Example
retriever = vector_store.as_retriever(search_kwargs={"k":2})

Now the retriever will return 2 chunks.

# 3.  invoke()

Purpose:
Send a query to the retriever and get documents.

Example:

docs = retriever.invoke("plant useful")

for doc in docs:
    print(doc.page_content)

This internally calls similarity search.

# Step 1: Vector store search
vector_store.similarity_search("plant useful")

# Step 2: Convert to retriever
retriever = vector_store.as_retriever()

# Step 3: Retrieve using retriever
docs = retriever.invoke("plant useful")

# Implementation 

```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS


loader = TextLoader(r"doc/data.txt")
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)

chunks = splitter.split_documents(docs)
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
vectorstore = FAISS.from_documents(chunks , embeddings)
vectorstore.add_documents(chunks)
retriever = vectorstore.as_retriever(search_type="mmr", search_kwargs={"k": 2})
```

```python
results = retriever.invoke("types of plant")
for i in results:
    print(i.page_content)
    print("-"*20)
```
